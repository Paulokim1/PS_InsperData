# <b>Análise Exploratória</b>

```python
data_ = data.drop_duplicates(subset = 'name')
```

```python
len(data_.name.unique()) == len(data.name.unique())
```

### <u>Análise Descritiva</u>

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(data_.describe().T, 
                 cmap = 'Blues', 
                annot = True, fmt = '.1f', 
                linewidths = 5, cbar = False, 
                annot_kws={"size": 16})
ax.set_title('Describe', fontsize = 18)
plt.xticks(size = 12)
plt.yticks(size = 12, rotation = 0);
```

### <u>Ranking</u>

```python
data_.sort_values(by = 'rank').head(10)
```

```python
data_.sort_values(by = 'favorites', ascending = False).head(10)
```

### <u>Visualização</u> 

```python
ax = sns.distplot(data_.favorites)
ax.figure.set_size_inches(12,6)
```

```python
ax = sns.boxplot(data = data_, 
            x = 'favorites')
```

```python
ax = sns.distplot(data_.popularity)
ax.figure.set_size_inches(12,6)
```

```python
ax = sns.boxplot(data = data_, 
            x = 'popularity')
```

```python
ax = sns.countplot(x = data_.genre, 
                   order = data_.genre.value_counts().index)
ax.figure.set_size_inches(30,6)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.show()
```

```python
grupoSource = data_.groupby(by = 'source')
count = grupoSource['episodes'].count().sort_values(ascending = False)

ax = sns.barplot(x = count.index, 
                 y = count.values)
ax.set_title("Num. Episódios VS Source")
ax.set_ylabel("Número de episódios")
ax.figure.set_size_inches(12,6)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.plot()
```

```python
plt.figure(figsize = (12,6))
ax = sns.heatmap(data_.corr(), 
                 cmap = 'Blues', 
                annot = True, fmt = '.1f', 
                linewidths = 5, cbar = False, 
                annot_kws={"size": 16})
ax.set_title('Correlações', fontsize = 18)
plt.xticks(size = 12)
plt.yticks(size = 12, rotation = 0);
```
